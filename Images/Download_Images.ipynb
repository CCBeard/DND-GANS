```python
import requests
import shutil
import pandas as pd
```

```python
url_file = pd.read_csv('Download_Scripts/dwarf_script.txt', sep='someobscurestring', header=None, engine='python')
url_list = url_file[0].to_numpy(dtype=str)
```

```python
print(len(url_list))
```

```python
for i in range(len(url_list)):
    url = url_list[i]
    file_name = '/Volumes/BigBoi/DND_Images/Dwarf/dwarf_{}.jpg'.format(i)
    
    res = requests.get(url, stream = True)
    
    if res.status_code == 200:
        with open(file_name,'wb') as f:
            shutil.copyfileobj(res.raw, f)
        print('Image sucessfully Downloaded: ',file_name)
    else:
        print('Image Couldn\'t be retrieved')
```
